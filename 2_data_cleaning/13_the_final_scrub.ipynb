```python
# Exercise 13: The Final Scrub

import pandas as pd
import numpy as np

patients = pd.DataFrame({
  'patient_id': [
    'PT-2001', 'PT-2001', 'PT-2002', 'PT-2003', 
    'PT-2004', 'PT-2005', 'PT-2006', 'PT-2006', 
    'PT-2007', 'PT-2008', 'PT-2009', 'PT-2010', 
    'PT-2011'
  ],
  'name': [
    'Anthony Ramirez', 'anthony ramirez', 'Grace Mitchell', 'Daniel Brooks', 
    'Nina Patel', 'Marcus Johnson', 'Elena Cruz', 'elena cruz', 
    'William Carter', 'Sophia Nguyen', None, 'Sophia Nguyen', 
    'Ethan Walker'
  ],
  'age': [
    '52', '52', 34, '28', 
    '61', 47, '26', 26, 
    '73', '39', 44, '39', 
    '58'
  ],
  'department': [
    'ER', 'er', 'Trauma', 'ER', 
    'Cardiology', 'Trauma', 'ER', 'er', 
    'ICU', 'ER', 'ER', 'ER', 
    'ICU'
  ],
  'triage_level': [
    'Critical', 'critical', 'Moderate', 'HIGH', 
    'Low', 'Moderate', 'critical', 'CRITICAL', 
    'High', 'Low', 'Moderate', 'low', 
    None
  ],
  'wait_time_minutes': [
    '12', '12', 55, '40 mins', 
    140, 60, '8', 8, 
    '95', '25', None, '25', 
    '70'
  ],
  'admitted': [
    True, True, False, True, 
    True, False, True, True, 
    False, True, False, True, 
    None
  ]
})

patients
```

```python
patients.dtypes

# Normalize text fields
patients['name'] = patients['name'].str.title()
patients['triage_level'] = patients['triage_level'].str.upper()
patients['department'] = patients['department'].str.upper()

# Convert numeric fields
patients['age'] = pd.to_numeric(patients['age'], errors='coerce')
patients['wait_time_minutes'] = (
    patients['wait_time_minutes']
    .astype(str)
    .str.replace(r'[^0-9]', '', regex=True)  # remove non-numeric
    .replace('', np.nan)                     # empty → NaN
    .astype(float)
)

# Remove duplicate records
patients = patients.drop_duplicates()

# Handle missing values
patients['name'] = patients['name'].fillna('Unknown')
patients['triage_level'] = patients['triage_level'].fillna('UNSPECIFIED')
patients['wait_time_minutes'] = patients['wait_time_minutes'].fillna(patients['wait_time_minutes'].median())
patients['admitted'] = patients['admitted'].fillna(False)

patients
```
