```python
# Exercise 12: Partiful Responses

import pandas as pd
import random

random.seed(1)

# Different possible RSVP variants
options = ['Yes', 'yes', 'yesss', 'Yep   ', ' yes', 'YES', 'y', 'No  ', 'no', 'N ', 'n', 'no thx']

one_hundred_responses = random.choices(options, k=100)

# Generate fake names for the 100 guests
names = [f'Guest_{i+1}' for i in range(100)]

# Create the DataFrame
responses = pd.DataFrame({
  'name': names,
  'rsvp': one_hundred_responses
})
responses
```

```python
# Standardize the capitalization
responses['rsvp'] = responses['rsvp'].str.lower()

# Strip extra spaces
responses['rsvp'] = responses['rsvp'].str.strip()

# Replace 'yes' and 'no' to 'y' and 'n' resp.
responses['rsvp'] = responses['rsvp'].str.replace('yes', 'y')
responses['rsvp'] = responses['rsvp'].str.replace('no', 'n')

responses
```
